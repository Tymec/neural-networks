```python
import matplotlib.pyplot as plt
import utils.activation_functions as af
from utils.helper import plot_function
```

```python
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3, figsize=(12, 8))
plot_function(af.identity, ax=ax1, label="Identity")
plot_function(af.binary_step, ax=ax2, label="Binary Step")
plot_function(af.sigmoid, ax=ax3, label="Sigmoid")
plot_function(af.relu, ax=ax4, label="ReLU")
plot_function(af.tanh, ax=ax5, label="Tanh")
plot_function(af.selu, ax=ax6, label="SELU", x_min=-3, x_max=3, y_min=-3, y_max=3)
plt.show()
```
